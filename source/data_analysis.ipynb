```python
X2 = X_new[['radius2', 'texture2', 'perimeter2', 'area2', 'smoothness2', 'compactness2', 'concavity2', 'concave_points2', 'symmetry2', 'fractal_dimension2']]
```

```python
import seaborn as sns  # 基于 Matplotlib 的高级绘图库
import matplotlib.pyplot as plt

# 设置画布大小
plt.figure(figsize=(15, 10))

# 遍历每一列并绘制密度图
for i, column in enumerate(X2.columns, 1):
    plt.subplot(4, 3, i)
    sns.kdeplot(X2[column], fill=True, color='orange')
    plt.title(f'Density of {column}')
    plt.xlabel('')
    plt.ylabel('')
    # 计算均值和标准差
    mean = X2[column].mean()
    std = X2[column].std()
    # 标记 3σ 范围
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.axvline(mean + 3*std, color='gray', linestyle=':', label='Mean + 3σ')
    plt.axvline(mean - 3*std, color='gray', linestyle=':', label='Mean - 3σ')
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(X2.columns, 1):
    plt.subplot(4, 3, i)
    plt.hist(X2[column], bins=30, color='skyblue', edgecolor='black')
    
    # 计算均值和标准差
    mean = X2[column].mean()
    std = X2[column].std()
    
    # 标记 3σ 范围
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.axvline(mean + 3*std, color='gray', linestyle=':', label='Mean + 3σ')
    plt.axvline(mean - 3*std, color='gray', linestyle=':', label='Mean - 3σ')
    
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
data_new = pd.concat([X1,X2,X3],axis=1)
data_new
```

```python
from sklearn.datasets import make_blobs
from sklearn import metrics
from sklearn.cluster import KMeans
```

```python
estimator = KMeans(n_clusters=2)
estimator.fit(data_new)
y = estimator.fit_predict(data_new)
y
```

```python
centroids = estimator.cluster_centers_
centroids
```

```python
metrics.calinski_harabasz_score(data_new, y)
```

```python
Y = X_new[['Diagnosis']]
result = pd.concat([data_new, pd.Series(y,index=data_new.index),Y],axis=1)
result.columns = list(data_new.columns)+['result']+['Diagnosis']
result
```

```python
result.to_csv('KMeans_result_.csv')
```
